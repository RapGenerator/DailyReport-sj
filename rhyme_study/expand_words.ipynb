```python
import fool
import json
import re
from pypinyin import lazy_pinyin, Style
from collections import Counter
```

```python
with open('words_d.json', 'r') as f:
    save_words = json.load(f)
words = save_words['words']
print(len(words)) # 715229
words
```

```python
yun = {
    'i': 'i',
    'u': 'u',
    'v': 'v',
    'a': 'a', 'ia': 'a', 'ua': 'a',
    'o': 'o', 'uo': 'o',
    'e': 'e', 'ie': 'e', 'ue': 'e', 've': 'e', 'er': 'e',
    'ai': 'ai', 'uai': 'ai',
    'ei': 'ei', 'uei': 'ei',
    'ao': 'ao', 'iao': 'ao',
    'ou': 'ou', 'iou': 'ou',
    'an': 'an', 'ian': 'an', 'uan': 'an', 'van': 'an',
    'en': 'en', 'uen': 'en', 'eng': 'en',
    'in': 'in', 'ing': 'in',
    'vn': 'vn',
    'ang': 'ang', 'iang': 'ang', 'uang': 'ang',
    'ong': 'ong', 'iong': 'ong'
}
```

```python
pos = set([p for _, p in words])
print(len(pos)) # 61
print(pos)
p_rm = set()
for pp in pos:
    p_word = set([w for w, p in words if p == pp])
    if len(p_word) < 10 or pp[0] == 'w':
        p_rm.add(pp)
print(p_rm)
pos = pos - p_rm
print(len(pos))
print(pos)
```

```python
word_dict = {}
bad_count = 0
for w, p in words:
    if p in pos:
        try:
            pinyin = [yun[tone] for tone in lazy_pinyin(w, style = Style.FINALS, strict=True)]
        except KeyError:
            bad_count += 1
#             print('KeyError', w, p)
        key = '_'.join(pinyin[::-1])
        if key not in word_dict:
            word_dict[key] = {p : {w : 1}}
        elif p not in word_dict[key]:
            word_dict[key][p] = {w : 1}
        elif w not in word_dict[key][p]:
            word_dict[key][p][w] = 1
        else:
            word_dict[key][p][w] += 1
print(bad_count) # 6833 if not judge in pos, 2301
word_dict
```

```python
with open('word_dict_c.json', 'w') as f:
    json.dump(word_dict, f)
```
