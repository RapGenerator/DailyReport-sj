```python
import pandas as pd
import numpy as np
import re
from collections import Counter
from pypinyin import lazy_pinyin, Style
```

```python
df = pd.read_csv('df_all.csv', header = -1, encoding = 'UTF-8', names = ['lyrics', 'is_bad'])
```

```python
lyrics = list(df.loc[:,'lyrics'])
lyrics
```

```python
target_lines = []
for line in lyrics:
    pinyin_list = lazy_pinyin(line, style = Style.FINALS, strict = True)
    if len(pinyin_list) >=2 and (pinyin_list[-2] == 'ong' or pinyin_list[-2] == 'iong') and pinyin_list[-1] == 'i':
        target_lines.append(line)
print(len(target_lines))
# target_lines
```

```python
yun = {
    'i': 'i',
    'u': 'u',
    'v': 'v',
    'a': 'a', 'ia': 'a', 'ua': 'a',
    'o': 'o', 'uo': 'o',
    'e': 'e', 'ie': 'e', 'ue': 'e', 've': 'e', 'er': 'e',
    'ai': 'ai', 'uai': 'ai',
    'ei': 'ei', 'uei': 'ei',
    'ao': 'ao', 'iao': 'ao',
    'ou': 'ou', 'iou': 'ou',
    'an': 'an', 'ian': 'an', 'uan': 'an', 'van': 'an',
    'en': 'en', 'uen': 'en', 'eng': 'en',
    'in': 'in', 'ing': 'in',
    'vn': 'vn',
    'ang': 'ang', 'iang': 'ang', 'uang': 'ang',
    'ong': 'ong', 'iong': 'ong'
}
```

```python
def get_target_lyrics(yuns : list):
    target_lines = []
    ya_len = len(yuns)
    for line in lyrics:
        pinyin_list = lazy_pinyin(line, style = Style.FINALS, strict = True)
        if ya_len <= len(pinyin_list):
            is_rhyme = True
            for i in range(1, ya_len+1):
                if yun.get(pinyin_list[-i], '') !=  yun[yuns[-i]]:
                    is_rhyme = False
                    break
            if is_rhyme:
                target_lines.append(line + '\n')
    return target_lines
```

```python
tl = get_target_lyrics(['ong', 'i'])
print(len(tl))
tl
```

```python
yun_names = set(yun.values())
len(yun_names)
```

```python
for y in yun_names:
    tl = get_target_lyrics([y])
    f = open(y + '.txt', 'w', encoding = 'UTF-8')
    f.writelines(tl)
    f.close()
    print(y + ' done')
```

```python
f = open('ong_i.txt', 'w', encoding = 'UTF-8')
f.writelines(tl)
f.close()
```

```python
for y2 in yun_names:
    for y1 in yun_names:
        tl = get_target_lyrics([y2, y1])
        f = open(y2 + '+' + y1 + '.txt', 'w', encoding = 'UTF-8')
        f.writelines(tl)
        f.close()
        print(y2 + '+' + y1 + ' done')
```

```python
for y3 in yun_names:
    for y2 in yun_names:
        for y1 in yun_names:
            tl = get_target_lyrics([y3, y2, y1])
            f = open(y3 + '+' + y2 + '+' + y1 + '.txt', 'w', encoding = 'UTF-8')
            f.writelines(tl)
            f.close()
            print(y3 + '+' + y2 + '+' + y1 + ' done')
```
